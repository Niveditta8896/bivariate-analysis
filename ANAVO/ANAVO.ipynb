# refer t test

```python
import pandas as pd
dataset=pd.read_csv("preplacement.csv")
```

```python
dataset
```

```python
from scipy.stats import ttest_ind
# step 2 :we r going to create 2 series of salaries for each gender 
M= dataset[dataset["gender"]=="M"]["salary"]
F=dataset[dataset["gender"]=="F"]["salary"]
ttest_ind(M,F)
```

```python
from scipy.stats import ttest_rel
male=dataset[dataset["gender"]=="M"]["ssc_p"]
male_2= dataset[dataset["gender"]=="M"]["hsc_p"]
ttest_rel(male,male_2)
```

```python
# note for the dependent t test we hv to test wheter the ssc_ mark and hsc_p mark is similar to the males. 
# fro the p value which is 49 percent we can say there is no difference btw them and it follows a null hypothesis cos in alternate
#hypothesis the p_value should be <0.05.

#hypothesis testing is about checking if 2 groups have a difference if there is p value is < 0.05.
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
sm.stats.anova_lm(ols('salary ~ C(specialisation) + C(workex) + C(specialisation):C(workex)', data=dataset).fit(), typ=2)

# first ols(ordinary least square) runs and finds the columnnames from data=dataset and finds the effect of specialisation on salary
#workex on salary and the effect of  interaction btw workex and specialisation on salary.

#then ols.fit all these findings gets saved to ols.
# then we run ANOVA (sm.stats.anova_lm)  on fitted ols and we run type 2 ANOVA(typ=2)

```

```python
# here p value for specialisation is 0.000587 so there is hig variance (p<0.05), specialisation significantly affects salry
#  here p value for workex is 0.000120 so there is hig variance (p<0.05), workex significantly affects salry
# here p value for C(specialisation):C(workex) is 0.576390 so there is significantly low variance  since p value is >0.05.
```
