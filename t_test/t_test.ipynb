# independent t test

```python
# y- to check similarity btw 2 different groups for eg salary in males and females
```

```python
from scipy.stats import ttest_ind
# step 2 :we r going to create 2 series of salaries for each gender 
M= dataset[dataset["gender"]=="M"]["salary"]
F=dataset[dataset["gender"]=="F"]["salary"]
ttest_ind(M,F)
```

```python
# now we r going to interpret the probability value:
# if p value  < 0.05 which means theres greater difference btw 2 groups and LOW SIMILARITY so there significant difference btw salarie
#of males and females. and T_value =2.1103433958668765 higher the value higher is the difference.
```

# dependent t test 

```python
# when do we do this test> when 2 groups r the same and we hv to check their similaritz based on a criteria.
```

```python
# how do we do: we r goin to check ssc_p and hsc_p marks for the males 
```

```python
from scipy.stats import ttest_rel
male=dataset[dataset["gender"]=="M"]["ssc_p"]
male_2= dataset[dataset["gender"]=="M"]["hsc_p"]
ttest_rel(male,male_2)
```

```python
# here t_value =0.6852881050898078
#p value =0.49431164600731503 means the probability is 40 percent so theres is very high similarity btw them .
```
