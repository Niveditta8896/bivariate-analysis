```python

quan=[]
qual=[]
def quan_qual(dataset):
    for columnname in dataset:
        if dataset[columnname].dtype=="O":
             qual.append(columnname)
        else :
            quan.append(columnname)
    return quan,qual # always if u want to return more than 2 outputs give return output1,output2, etc.
    
```

```python
def vif(dataset):
    import numpy as np
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    
    
    values=dataset.values
    vif=[variance_inflation_factor(values,i)for i in range(values.shape[1])]
    vif_table=pd.DataFrame({"vif_values":vif,"variables":dataset.columns})
    return vif_table
```

```python
#here vif for salary is less than 5 so multicollinearity is less for salary with respect to others. for others are above 10.
```
